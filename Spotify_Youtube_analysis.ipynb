#### Q4. How do the average views, likes, and comments are compared between different album types?

```python
df_mean = data.groupby('Album_type')[['Views', 'Likes', 'Comments']].mean()
df_mean
```

```python
df_mean = df_mean.reset_index()
df_mean
```

```python
#using melt function to unpivot the data frame
df_mean_melted = pd.melt(df_mean, id_vars = 'Album_type', var_name = 'Attribute', value_name = 'Total')
df_mean_melted
```

```python
# 1. Create the figure/axes on which to draw the plot
plt.figure(figsize = (9,4)) 

# 2. Draw ONE bar plot.
# This plot will have bars grouped by 'Album_type' and colored by 'Attribute'.
sns.barplot(
    x = 'Album_type', 
    y = 'Total', 
    hue = 'Attribute', 
    data = df_mean_melted
)

# 3. Display the plot
plt.show()
```
